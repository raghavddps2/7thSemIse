```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import preprocessing
%matplotlib inline
```

```python
df = pd.read_csv('CustomerData.csv')
df.head()
df.gender.fillna(df.gender.mode()[0],inplace=True)
df.age.fillna(int(df.age.mean()),inplace=True)
df["annual income (lakhs)"].fillna(df["annual income (lakhs)"].mean(),inplace=True)
df.category.fillna(df.category.mode()[0],inplace=True)
```

```python
sns.distplot(df["age"])
plt.show()
```

```python
sns.boxplot(df["age"])
plt.show()
```

```python
#Removing outliers as people in this range do not have stable earnings
df.drop(df[df["age"] < 17].index,inplace=True)
```

```python
sns.distplot(df["age"])
plt.plot()
```

```python
sns.scatterplot(data=df,x=df.age,y="annual income (lakhs)")
```
