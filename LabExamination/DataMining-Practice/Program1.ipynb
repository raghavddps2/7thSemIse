```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df[df['Age'] > 100] = np.nan
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(['CustID'],axis=1)
```

```python
df.head()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isnull().sum()
```

```python
df.tail()
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df.tail()
```

```python
df['Income'] = df['Income'].fillna(df['Income'].mean())
```

```python
df['Income'].isnull().sum()
```

```python
scaler = MinMaxScaler()
```

```python
df[['Income']] = scaler.fit_transform(df[['Income']])
```

```python
df.head()
```

```python
bins = [10,30,50,70,100]
labels = ['Child','Adult','Uncle','Namaste dadaji']
df['CustCategory'] = pd.cut(df['Age'],labels=labels,bins=bins)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Mode'] = df['Mode'].fillna(df['Mode'].mode()[0])
```
