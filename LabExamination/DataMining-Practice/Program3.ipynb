```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from apyori import apriori
```

```python
df = pd.read_csv('transactions.csv')
```

```python
df.head()
```

```python
df = df.drop(['TID'],axis=1)
```

```python
df.head()
```

```python
records = []
for i in range(len(df)):
    record = []
    for j in range(len(df.columns)):
        if(df.values[i,j]):
            record.append(df.columns[j])
    records.append(record)
```

```python
records
```

```python
min_support = 0.03
min_confidence = 0.7
rules = list(apriori(records,min_support=min_support,min_confidence=min_confidence))
```

```python
df1 = pd.DataFrame(rules)
```
