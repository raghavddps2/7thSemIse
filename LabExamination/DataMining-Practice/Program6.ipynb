```python
from sklearn.cluster import DBSCAN,KMeans,AgglomerativeClustering
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
X,Y = make_moons(250,noise=0.025,random_state=22)
plt.scatter(X[:,0],X[:,1])
```

```python
# Applying KMeans algorithm
labels = KMeans(n_clusters=2).fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=labels,cmap='cividis')
```

```python
# Applying DBSCAN Algorithm
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
# Showing the DBSCAN
dbscan = DBSCAN(eps=0.2)
labels = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
```

```python
import pandas as pd
df = pd.DataFrame(X)
df.to_csv('make_moons.csv')
```

```python
agg = AgglomerativeClustering(linkage='complete').fit(X)
print(agg.__doc__)
labels = agg.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
```
